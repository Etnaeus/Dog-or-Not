```python
#Dataset from https://www.kaggle.com/code/adinishad/pytorch-cats-and-dogs-classification
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_dir = './Dataset/training_set'
test_dir = './Dataset/test_set'
```

# Normalisation

```python
#https://www.geeksforgeeks.org/python/how-to-normalize-center-and-standardize-image-pixels-in-keras/
datagen = ImageDataGenerator(validation_split=0.2, #20% of data as validation
                             rescale=1.0/255)

train_generator = datagen.flow_from_directory(train_dir,
                                              target_size=(300,300), # size of image
                                              color_mode='grayscale', # grayscale the images
                                              class_mode='binary') # 2 classes

validation_generator = datagen.flow_from_directory(train_dir,
                                                   target_size=(300,300),
                                                   color_mode='grayscale',
                                                   class_mode='binary')
```
